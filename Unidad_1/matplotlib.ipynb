```python
import matplotlib as plt
import matplotlib.pyplot as plt
```

```python
# Valores para  la gráfica
x_values = [0, 1, 2, 3, 4, 5]
y_values = [0, 1, 4, 9, 16, 25]

# Crear la gráfica
plt.plot(x_values, y_values)

# Mostrar la gráfica
plt.show()
```

```python
plt.plot([1, 2, 3, 4])
plt.ylabel('some numbers')
plt.show()
```

```python
y1 = []
y2 = []
x = range(-100, 100, 10)
for i in x:
    y1.append(i**2)
    y2.append(-i**2)

plt.plot(x, y1)
plt.plot(x, y2)
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-2000, 2000)
plt.axhline(0)
plt.axvline(0)
plt.show()
```

```python
x = [1, 2, 3, 4]
y = [2, 4, 1, 3]
plt.plot(x, y, label='First Line')
plt.xlabel('eje x')
plt.ylabel('eje y')
plt.title('Interesante')
plt.show
```

```python
x1 = [1, 2, 3, 4]
y1 = [2, 4, 3, 1]
x2 = [1, 2, 3, 4]
y2 = [4, 1, 3, 2]
plt.plot(x1, y1, label='First Line')
plt.plot(x2, y2, label='Second Line')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Interesante')
plt.show()
```

```python
# Gráfico de líneas y puntos
x = [1, 2, 3, 4, 5, 6]
y = [5, 2, 4, 2, 1, 4]

# Graficando los puntos
plt.plot(x, y, color='red', linestyle='dashed', linewidth = 4, marker='o', markerfacecolor='black', markersize=12)

# Etiquetas de los ejes y título
plt.xlabel('Eje x')
plt.ylabel('Eje y')
plt.title('Gráfico de líneas y puntos')
```

```python
# gráfica de barras

valores = [5, 6, 3, 4, 6]
nombres = ['A', 'B', 'C', 'D', 'E']

plt.bar(nombres, valores, color='purple')
plt.show()
```

```python
# gráfico de barras horizontales
valores = [5, 6, 3, 4, 6]
nombres = ['A', 'B', 'C', 'D', 'E']

plt.barh(nombres, valores, color='yellow')
plt.show()
```

```python
# gráfico de barras apiladas y de diferentes colores

# altura de las barras
altura = [3, 12, 5, 18, 9]

# etiquetas de las barras
etiquetas = ['A', 'B', 'C', 'D', 'E']

# Graficando
c1 = ['red', 'green', 'blue', 'yellow', 'purple']
c2 = ['one', 'two', 'three', 'four', 'five']
plt.bar(etiquetas, altura, width=1, color=c1)

# Etiquetas de los ejes y título
plt.xlabel('Eje x')
plt.ylabel('Eje y')
plt.title('Gráfico de barras apiladas')
plt.show()
```

```python
# Histograma

# Frencuencia de los datos
edades = [22, 55, 62, 45, 21, 22, 34, 42, 42, 4, 2, 102, 95, 85, 55, 110, 120, 70, 65, 55, 111, 115, 80, 75, 65, 54, 44, 43, 42, 48]

# Rango de edades e intervalos
rango = (0, 120)
intervalos = 10

# Graficando
plt.hist(edades, intervalos, rango, color='green', histtype='bar', rwidth=0.8)

# Etiquetas de los ejes y título
plt.xlabel('Edades')
plt.ylabel('Frecuencia')
plt.title('Histograma')
plt.show()
```

```python
# Puntos dispersos

# Coordenadas de los puntos
x = [1, 2, 3, 4, 5, 6, 7, 8]
y = [5, 2, 4, 2, 1, 4, 5, 2]

# Graficando
plt.scatter(x, y, label='Puntos', color='red', s=100, marker='x') 
plt.legend()
plt.show()
```

```python
# Gráfico de pastel

# Etiqutas de las secciones
secciones = ['A', 'B', 'C', 'D']

# Colores de cada sección
colores = ['c', 'm', 'r', 'b']

# Tamaño de cada sección
tamanos = [25, 35, 20, 20]

# Graficando
plt.pie(tamanos, colors=colores, labels=secciones, shadow=True, startangle=45, explode=(0.1, 0.1, 0.1, 0.1), 
        autopct='%1.1f%%', radius=1.5)
plt.axis('equal')
plt.legend(title='Secciones')
plt.show()
```

```python
# graficar curvas de una ecuación

import numpy as np
# coordenadas x
x = np.arange(0, 5*(np.pi), 0.1)
# coordenadas y
y = np.sin(x)
# graficando   
plt.plot(x, y)
plt.show()
```
